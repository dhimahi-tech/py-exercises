##### Problem 4:

There is a logical bug in the below problem. The enhanced sky color has contrasting colors (grey, green, red etc) in comparison to the original image color. Can you help analyze this defect and fix it.

> Original Image Credit: https://www.pexels.com/photo/brown-field-and-blue-sky-46160/

```python
# Important: Run this cell before moving to execute the next cell.
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline

cloud_image = '../images/clouds.jpg'
```

```python
# Find the defect, analyze and fix this code in the following cell.
# Pre-condition: Run the above cell prior to this as it has imports to OpenCV & other libraries

img = cv2.imread(cloud_image, cv2.IMREAD_COLOR)

# Array with a 35% scale factor of the original image.
scaled_array = np.ones(img.shape) * 1.35

# Perform matrix multiplication with scaled image and fetch a unsigned 8bit array.
img_enh = np.uint8(cv2.multiply(np.float64(img), scaled_array))

# Display the images.
plt.figure(figsize=(10, 5))
plt.subplot(121); plt.imshow(img[:, :, ::-1]);     plt.title('Original Imag')
plt.subplot(122); plt.imshow(img_enh[:, :, ::-1]); plt.title('With 35% enh')
```
